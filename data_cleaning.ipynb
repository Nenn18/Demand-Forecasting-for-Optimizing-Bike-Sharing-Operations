# **前処理**

## **train/test**

```python
#trainデータとtestデータに分ける
train = status[status["predict"] == 0]
test = status[status["predict"] == 1]

#不要なカラムを削除
train = train.drop(columns = "predict")
test = test.drop(columns = ["predict", "bikes_available"])
```

```python
#「年」「月」「日」を結合して、datetime型に変形
train["date"] = pd.to_datetime(
    train["year"].astype(str) + "-" +
    train["month"].astype(str).str.zfill(2) + "-" +
    train["day"].astype(str).str.zfill(2)
)
train["date"] = pd.to_datetime(train["date"]).dt.date
train["date"] = pd.to_datetime(train["date"])

test["date"] = pd.to_datetime(
    test["year"].astype(str) + "-" +
    test["month"].astype(str).str.zfill(2) + "-" +
    test["day"].astype(str).str.zfill(2)
)
test["date"] = pd.to_datetime(test["date"]).dt.date
test["date"] = pd.to_datetime(test["date"])

#平日と土日でダミー化
train["weekday"] = (train["date"].dt.weekday >= 5).astype(int)

test["weekday"] = (test["date"].dt.weekday >= 5).astype(int)
```

# **データの結合**

train/testデータに以下のデータのカラムを結合

stationデータ："station_id", "dock_count", "city", "installation_date"

weatherデータ："date", "events"

```python
train = pd.merge(train, station[["station_id", "dock_count", "city", "installation_date"]], on = "station_id", how = "left")
test = pd.merge(test, station[["station_id", "dock_count", "city", "installation_date"]], on = "station_id", how = "left")
```

```python
train = pd.merge(train, weather[["date", "events"]], on = "date", how = "left")
test = pd.merge(test, weather[["date", "events"]], on = "date", how = "left")
```

```python
#天気のカラムは、雨が降っているか降っていないかでダミー化
train["events"] = (train["events"].isnull()).astype(int)
test["events"] = (test["events"].isnull()).astype(int)
```

```python
#"bikes_available"が欠けているレコードは、学習データに含めない。
train = train[train["bikes_available"].notnull()]
```

```python
#csvで保存
train.to_csv("/content/drive/My Drive/Python/SIGNATE/bike/cleaning_train.csv", index = False)
test.to_csv("/content/drive/My Drive/Python/SIGNATE/bike/cleaning_test.csv", index = False)
```
